#### Q3. 
Crie um dataframe `participacao` com as colunas `pa√≠s`, `copa`, e `num_copas` que armazena a quantidade de copas do mundo que cada pa√≠s participou, informando se √© da copa masculina ou feminina. Ao final imprima, como apresentado a seguir, o top 5 pa√≠ses de cada competi√ß√£o que mais participou de copas do mundo.

Exemplo de sa√≠da (valores inventados):
```
pa√≠s            copa        num_copas

```

```

Brazil          Masculina   21
```

````python
# Criar os dataframes separados para time_1 e time_2
time_1_df = wc[['time_1', 'ano', 'copa']].rename(columns={'time_1': 'pa√≠s'})
time_2_df = wc[['time_2', 'ano', 'copa']].rename(columns={'time_2': 'pa√≠s'}) 

# Combinar os dataframes e remover duplicatas
participacao_df = pd.concat([time_1_df, time_2_df])
participacao_df = participacao_df.drop_duplicates(subset=['pa√≠s', 'ano', 'copa'])

# Contar o n√∫mero de participa√ß√µes por pa√≠s e tipo de competi√ß√£o
participacao = participacao_df.groupby(['pa√≠s', 'copa']).size().reset_index(name='num_copas')

# Separar participa√ß√µes por tipo de competi√ß√£o
participacao_masculina = participacao[participacao['copa'] == 'Masculina']
participacao_feminina = participacao[participacao['copa'] == 'Feminina']

# Obter o top 5 pa√≠ses de cada competi√ß√£o
top5_masculina = participacao_masculina.sort_values(by='num_copas', ascending=False).head(5)
top5_feminina = participacao_feminina.sort_values(by='num_copas', ascending=False).head(5)

# Exibir os resultados
display(top5_masculina)
display(top5_feminina)
````

#### Q5. 
Qual pa√≠s tomou mais cart√µes amarelos somando todas as copas?

Neste exerc√≠cio voc√™ vai trabalhar com as colunas `cartao_amarelo_1` e `cartao_amarelo_2` onde cada observa√ß√£o √© uma string represetando uma lista dos cart√µes amarelos de um √∫nico jogo na forma `minuto|placar|jogador(a)`. Por exemplo:
```
['16‚Äô|1:0|Rosana G√≥mez', '20‚Äô|2:0|Gabriela Ch√°vez]
```

Recomendo criar colunas `num_cartoes_1` e `num_cartoes_2` a partir de cada coluna `cartao_amarelo_X` usando o m√©todo gen√©rico `apply` para chamar uma fun√ß√£o que voc√™ vai criar para transformar uma observa√ß√£o de cart√£o amarelo em uma contagem de elementos da lista. 

Lembre de levar em considera√ß√£o que muitas observa√ß√µes s√£o `NaN`. 

Em seguida fa√ßa sua m√°gica para agrupar as informa√ß√µes por pa√≠s, acumular os cart√µes de jogos em casa e jogos visitante, e produzir o resultado final como apresentado a seguir (valores inventados):

```
pa√≠s        cart√µes amarelos
Argentina   97
England     93
Australia   93
...
```

````python
def contar_cartoes(cartoes_str):

    #retornar se for NaN
    if pd.isna(cartoes_str):
        return 0
    
    #remover caracteres indesejados
    cartoes_str = cartoes_str.replace('&rsquor;', "'").replace("[", "").replace("]", "").replace("'", "")

    cartoes_str = cartoes_str.split(", ") #dividir cart√µes em forma de lista
    return len(cartoes_str) #retornar contagem de cart√µes

cartoes_1_df = wc[['cartao_amarelo_1', 'time_1']].rename(columns={'time_1':'pais','cartao_amarelo_1': 'cartoes amarelos'})
cartoes_2_df = wc[['cartao_amarelo_2', 'time_2']].rename(columns={'time_2':'pais','cartao_amarelo_2': 'cartoes amarelos'})

cartoes_df = pd.concat([cartoes_1_df, cartoes_2_df])

#aplicar fun√ß√£o na coluna de cart√µes amarelos
cartoes_df['cartoes amarelos'] = cartoes_df['cartoes amarelos'].apply(contar_cartoes)

#fazer novo dataframe
cartoes_amarelos = cartoes_df.groupby(['pais'])['cartoes amarelos'].sum().reset_index(name='cart√µes amarelos')
top = cartoes_amarelos.sort_values(by='cart√µes amarelos', ascending=False)

display(top)
````

#### Q6.
Qual o top10 jogadores com mais gols em copa? Considere gols em jogo e gols de p√™nalti.

Para conseguir essa informa√ß√£o, voc√™ precisar√° trabalhar com as colunas:
```
10  gols_1_detalhes         
11  gols_2_detalhes    
14  gols_1_penalti     
15  gols_2_penalti     
```

Essas tamb√©m s√£o colunas textuais, onde cada observa√ß√µes apresenta todos os gols de uma partida separados pelo caracter `|`. Cada gol est√° na forma `'Jogador(a) ¬∑ minuto‚Äô'`. Por exemplo:
```
'Alex Morgan ¬∑ 12‚Äô|Rose Lavelle ¬∑ 20‚Äô'
```

Lembre de levar em considera√ß√£o que muitas observa√ß√µes s√£o `NaN`. 

Recomendo criar um dicion√°rio √† parte, onde cada chave ser√° um jogador encontrado nessas colunas do dataframe `wc` e o valor correspondente ser√° a contagem de ocorr√™ncias desses nomes.

Em seguida basta converter o seu dicion√°rio em um dataframe e imprim√≠-lo na forma (valores n√£o s√£o inventados üòÅ):
```
jogador(a)      num_gols 
Marta           17
Miroslav Klose  16
... 
```

````python
jogadores_e_gols = {}

def Main():

    #renomear todos para apenas gols
    gols_1_detalhes_df = wc[['gols_1_detalhes']].rename(columns={'gols_1_detalhes':'gols'})
    gols_2_detalhes_df = wc[['gols_2_detalhes']].rename(columns={'gols_2_detalhes':'gols'})
    gols_1_penalti_df = wc[['gols_1_penalti']].rename(columns={'gols_1_penalti':'gols'})
    gols_2_penalti_df = wc[['gols_2_penalti']].rename(columns={'gols_2_penalti':'gols'})

    #concatenar tudo em um s√≥ dataframe
    gols_df = pd.concat([gols_1_detalhes_df, gols_2_detalhes_df, gols_1_penalti_df, gols_2_penalti_df])
    
    #aplicar fun√ß√£o em todos os itens
    gols_df['gols'].apply(processar_gols)

    #criar dataframe com resultado
    gols_df = criar_dataframe_jogadores()

    #define os 10 melhores jogadores
    top_10_jogadores = gols_df.sort_values(by='numero_de_gols', ascending=False).head(10)

    #mostrar na tela
    display(top_10_jogadores)

def processar_gols(gols_str):
    '''Divide os gols de cada partida'''
    #pular NaN
    if pd.isna(gols_str):
        return 0
    
    #dividir jogadores
    gols_str = gols_str.split('|')

    for item in gols_str:

        #deixar s√≥ o nome do jogador
        item_tmp = item
        if '¬∑' in item:
            item_tmp = item.split(' ¬∑ ')
            item_tmp = item_tmp[0]
        
        adicionar_ao_dicionario(item_tmp)
    
def adicionar_ao_dicionario(item):
    '''adiciona jogadores novos e atualiza os j√° existentes no dicion√°rio'''
    if item in jogadores_e_gols:
        jogadores_e_gols[item] += 1
    else:
        jogadores_e_gols[item] = 1
    
def criar_dataframe_jogadores():
    '''Cria um novo dataframe com um dicion√°rio de jogadores/gols'''

    df_jogadores = pd.DataFrame(list(jogadores_e_gols.items()), columns=['jogador', 'numero_de_gols'])

    return df_jogadores
Main()
````
